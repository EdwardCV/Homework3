#Set up

```python
import pandas as pd
survey = pd.read_csv('/content/surveydataFall2024.csv')
survey
```

#A) Show only rows where Statistics skill is 10

```python
survey.query('Statistics == 10')
```

#B) Show only rows where both Statistics and Math skills are 10

```python
survey.query('Statistics == 10 and Math == 10')
```

#C) Show only the Semester, Communication, and Visualization columns where both Statistics and Math skills are 10

```python
survey.query('Statistics == 10 and Math == 10').filter(['Semester','Communication','Visualization'])
```

#D) Show only the Semester, Communication and Visualization columns in decreasing order of Statistics and then Math skills.

```python
survey.sort_values(['Statistics','Math'],ascending=False).filter(['Semester','Communication','Visualization'])
```

#E) Show only rows where Statistics skill is missing.

```python
survey.query('not Statistics > 0')
```

#F) Show for every student, only their Semester, Math skill, Statistics skill, and the maximum of their Math and Statistics skills (Hint: use axis=1)

```python
survey.filter(['Semester','Math','Statistics']).assign(Maximum=survey[['Math','Statistics']].max(axis=1))
```

#G) Show the median value of Computer Science Skills.

```python
survey.filter(['ComputerScience']).median()
```

#H) Show the median value of Computer Science skills in Fall 2023.

```python
survey.query('Semester == "Fall2023"').filter(['ComputerScience']).median()
```

#I) Show the median values of (each of) Computer Science, Math, and Statistics skills in every semester

```python
survey.columns.drop(['Semester'])
```

```python
(
    survey
    .filter(['Semester','ComputerScience','Math','Statistics','TakenCPSC483'])
    .melt(id_vars=['Semester'],value_vars=['ComputerScience','Math','Statistics'],var_name='Skill',value_name='Value')
    .dropna()
    .groupby(['Semester','Skill'])[['Value']]
    .median()
)
```

#J) Show the median values of (each of) Computer Science, Math, and Statistics skills in every semester of students who have taken CPSC483 (Taken CPSC483 is Yes)

```python
(
    survey
    .filter(['Semester','ComputerScience','Math','Statistics','TakenCPSC483'])
    .query('TakenCPSC483 == "Yes"')
    .melt(id_vars=['Semester'],value_vars=['ComputerScience','Math','Statistics'],var_name='Skill',value_name='Value')
    .dropna()
    .groupby(['Semester','Skill'])[['Value']]
    .median()
)
```

#K) Show the median values of (each of ) Computer Science, Math, and Statistics skills for groups of students who have taken and not taken CPSC483 (Hint: groupby)

```python
(
    survey
    .filter(['ComputerScience','Math','Statistics','TakenCPSC483'])
    .melt(id_vars=['TakenCPSC483'],value_vars=['ComputerScience','Math','Statistics'],var_name='Skill',value_name='Value')
    .dropna()
    .groupby(['TakenCPSC483','Skill'])[['Value']]
    .median()
)
```
